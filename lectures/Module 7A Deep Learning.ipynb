```python
## import some useful modules
import tensorflow as tf
import numpy as np
from tensorflow import keras
```

```python
## create two arrays, x and y.
x = np.random.randn(1000)
y = 3*x+2
```

- certainly we know that y = 3*x + 2. 
- Now, we feed the machines x and y. 
- The machine does not know the relationship. 
- we will use deep learning to teach the machine to learn the relationship

```python
x.shape    
# x is a 1000*x vector
# it contains only 1 feature and 1000 samples
# so the input shape to the neural network is 1
```

```python
## a simple neural network
model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
## we create a deep learning model.
```

### Explanation of the above neural network
- the core building blocks of a neural network are layers
- classify the network as a sqeuential model using tf.keras.Sequential(). Most of the models we learn are sequential
- the above network has just one layer apart from the input layer.
- the layer is "dense". It means the layer learns by looking at the global pattern of input data (fully connected). Later, we will also look at convolution layers, which look at local patterns of input data.
- for the first hidden layer, you need to specify the shape of the input array, input_shape=[1]. You just need to specify the number of features, not the sample size.



```python
# after building layers, we need to compile the model
model.compile(optimizer='sgd', loss='mean_squared_error')
```

### compile a model
- After building layers, we need to compile the model
    1. specify the optimizer:  the algorithm which the model uses to adjust weight for minimizing loss value. The optimizer here is sgd (stochastic gradient descent)
    2. loss, the loss function used to calculate the loss value. (the loss function is "mean_squared_error"

```python
# now, we can train the model by using model.fit()
model.fit(x, y, epochs=500)      # number of epochs is related to number of iterations
```

```python
# after the model is trained, we can see if the model can predict an accurate value of y.
# we feed the model with x = 10, and the true value of y = 3*10 +2 = 32
model.predict([10])
```

```python
# Example 2: Classify handwritten digits
# download MNIST dataset from keras.datasets
# split the dataset into training and test sets
import tensorflow as tf
mnist = tf.keras.datasets.mnist
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()
# mnist is built-in dataset in tensorflow. 
# Tensorflow has a lot of built-in datasets to work with.
# we spilt the data into two sets, the training set and the test set.
```

```python
train_images.shape
# there are 60,000 images
# each image is represented by 28*28 2-D tensor (2-D numpy array)
```

```python
train_labels
# a 1-D tensor (vector) of 60,000 elements
# each element indicates the class of the image
```

```python
import matplotlib.pyplot as plt
plt.imshow(train_images[0], cmap="binary")
print(train_labels[0])

# show the 0th train image, which is a number 5.
```

```python
# pre-process the data.
# the value of each element is between 0 and 255, 
# the training will be easier if all the values are convereted between 0 and 1. 
# this process is called normalizing.
# we do the same for test_image

train_images = train_images.astype('float32') / 255


test_images = test_images.astype('float32') / 255
```

```python
# now we can build the model
model = tf.keras.models.Sequential([tf.keras.layers.Flatten(input_shape=(28,28)),   # input layer
                                    tf.keras.layers.Dense(128, activation=tf.nn.relu),   # first hidden layer
                                    tf.keras.layers.Dense(10, activation=tf.nn.softmax)]) # output layer
```

### Explanation of the above neural network
- the core building blocks of a neural network are layers
- classify the network as a sqeuential model using models.Sequential(). Most of the models we learn are sequential
- after that we can add layers to the model
- the above model has 3 layers
    1. the input layer
    2. the first hidden layer
    3. the output layer
- the input layer. This layer first flattens the input array. You don't need to specify the sample size. This layer is not necessary if you already flattened the input array.
- the first hidden layer:
    1. the layer is "dense". It means the layer learns by looking at the global pattern input data (fully connected). Later, we will also look at convolution layers, which look at local patterns of input data.
    2. for the first hidden layer, the shape of the output is 128. It transforms the shape of the output from 28*28 to 128.
    3. activation function tells what tensor operation is done on the input data.
    4. Relu effectively means "If X>0 return X, else return 0" -- so what it does it it only passes values 0 or greater to the next layer in the network.

- the output layer
    1. The output from the previous layer is fed into this layer. Again, the layer is dense.
    2. the shape of the output is 10, the data is further transformed from 128 features to 10 features.
    3. activation function is Softmax: 
    4. Softmax takes a set of values, and effectively picks the biggest one, so, for example, if the output of the last layer looks like [0.1, 0.1, 0.05, 0.1, 9.5, 0.1, 0.05, 0.05, 0.05], it saves you from fishing through it looking for the biggest value, and turns it into [0,0,0,0,1,0,0,0,0] -- The goal is to save a lot of coding!
    5. softmax activation function is commonly used in the output layer of a network for a classification task with 3 or more classes.

### relu function


```python
# then we compile the model.
model.compile(optimizer = tf.optimizers.Adam(),
              loss = 'sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

### Compile the model
- After setting up the model, we need to compile it.
-  we need to specify 
    1. optimizer (mechanism to minimize loss values): Adam(). a very commonly used optimizer.
    2. loss, the loss function used to calculate the loss value: loss function here is "sparse_categorical_crossentropy". this is commonly used for classification task
    3. metrics, evaluate the performance of the model
- loss function and metrics are different
    1. loss function: a function to calculate the loss value for the machine to use. The machine uses loss value to do the learning.
    2. metrics: a measure which humans use to evaluate the model 

```python
# finall, we can fit the model.
model.fit(train_images, train_labels, epochs=5)
# epochs=5
# iterate on the training data 
# after each epoch, we can see the loss value and accuracy value on training data.
# loss value decreases over epochs, accuracy value increases over epochs.
```

```python
model.evaluate(test_images, test_labels)
# now we can check the performance of the model on test set.
# the first number is the loss score on test data.
# the second number is the accuracy of the model in recognizing the handwritten digits in the test set
```

### Preparing the data
- the input data is lists of integer, and the lists have different lengths. This input data cannot be fed into a nueral network.
- we need to put our lists into one numpy array to feed into the nueral network.
- we will one-hot encode our lists to turn them into vectors of 0s and 1s.
- E.g., suppose there is a review with just two words, it is represented as a list [3, 5]. We can convert this list into a 10,000-dimensional array that would be all 0s except for indices 3 and 5, which would be 1s. 
- Given that there are 25000 review in the training set. The input data data will be converted to a 25,000x10,000 numpy array.

```python
import numpy as np

def vectorize_sequences(sequences, dimension=10000):
    results = np.zeros((len(sequences), dimension))     # 1.        
    for i, sequence in enumerate(sequences):
        results[i, sequence] = 1.                    #2.          
    return results

x_train = vectorize_sequences(train_data)                  
x_test = vectorize_sequences(test_data)   

# 1. Creates an all-zero matrix of shape (len(sequences), dimension)

# 2. Sets specific indices of results[i] to 1s

# We do the same for both train_data and test_data
```

```python
x_train[0]
# the first review is now converted into a vector 10000 elements.
# each element is either 0 or 1.
```

```python
# train_labels is a list of numbers, we need to convert it into numpy array.
# we do the same for test_labels.
y_train = np.asarray(train_labels)
y_test = np.asarray(test_labels)
```

```python
y_train
```

```python
# we can build our newtork.
model = tf.keras.models.Sequential([tf.keras.layers.Dense(16, activation=tf.nn.relu, input_shape=(10000,)), # first hidden layer
                                    tf.keras.layers.Dense(16, activation=tf.nn.relu),  # second hidden layer
                                    tf.keras.layers.Dense(1, activation=tf.nn.sigmoid)]) # output layer
```

### Explanation of the network.
- all the layers are dense layers.
- In the first hidden layer, the shape of each input data point is 10000. the shape of each output data point is 16. We use relu activation function.
- In the second layer, the shape of each output data point is 16. We use relu activation function.
- the final layer uses a sigmoid activation so as to output a probability (a score between 0 and 1, indicating how likely the sample is to have the target “1”: how likely the review is to be positive)
- sigmoid activation function is commonly used for the output layer of the newtwork for binary classification

### Sigmoid function


```python
# we can now compile the model
model.compile(optimizer='rmsprop',
              loss='binary_crossentropy',
              metrics=['accuracy'])

# optimizer (the mechanism to minimize loss values): rmsprop
# loss='binary_crossentropy', one of the best loss functions for binary classification problem.
```

```python
# we can now fit the model.
history = model.fit(x_train,
                    y_train,
                    epochs=20,
                    batch_size=512,
                    validation_split = 0.2)


# We split the input data set into two sets: a training set and the validation set.
# the size of validation set = 0.2 of the input data set.
# we train the model using batches. The size of each batch = 512.
```

```python
# model.fit() returns a History object. 
# This object has a member history, which is a dictionary containing data about everything that happened during training.
history_dict = history.history
history_dict.keys()
```

```python
# we can plot the changes of training loss and validation loss over epochs.

import matplotlib.pyplot as plt

history_dict = history.history
loss_values = history_dict['loss']
val_loss_values = history_dict['val_loss']

epochs = range(1, len(loss_values) + 1)

plt.plot(epochs, loss_values, 'bo', label='Training loss')           
plt.plot(epochs, val_loss_values, 'b', label='Validation loss')      
plt.title('Training and validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.show()
```

```python
# we can plot the changes of training accuracy and validation accuracy over epochs.

acc = history_dict['accuracy']
val_acc = history_dict['val_accuracy']

plt.plot(epochs, acc, 'bo', label='Training acc')
plt.plot(epochs, val_acc, 'b', label='Validation acc')
plt.title('Training and validation accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

plt.show()
```

### changes in loss and accuracy over epochs.
- the training loss decreases with every epoch, and the training accuracy increases with every epoch.
- validation loss first decreases and then starts to rise with epoch.
- validation accuracy first increases and then starts to falls with epoch.
- so if we run the models with too many epochs, there will be a overfitting problem.
- based on the above graphs, the good number of epoch is 4.
- Such analysis is not only used to choose the best number of epoch. It is also used to tune other hyperparameters

```python
# we start from scratch and train the model with the entire input data set.
# the number of epochs = 4.
model = tf.keras.models.Sequential([tf.keras.layers.Dense(16, activation=tf.nn.relu, input_shape=(10000,)), # first hidden layer
                                    tf.keras.layers.Dense(16, activation=tf.nn.relu),  # second hidden layer
                                    tf.keras.layers.Dense(1, activation=tf.nn.sigmoid)]) # output layer


model.compile(optimizer='rmsprop',
              loss='binary_crossentropy',
              metrics=['accuracy'])

model.fit(x_train, y_train, epochs=4, batch_size=512)
```

```python
# finall, we run the model on test data.
results = model.evaluate(x_test, y_test)
results
```

```python
# Example 4: predict housing price
# this is a regression problem, not a classification task
# first, load the dataset.

boston_housing = tf.keras.datasets.boston_housing

(train_data, train_targets), (test_data, test_targets) = boston_housing.load_data()
```

```python
train_data.shape
# theare are 404 training samples, each with 13 numerical features, 
```

```python
train_targets
# The targets are the median values of owner-occupied homes, in thousands of dollars
```

```python
# the data are numpy arrays, which can be fed into nueral networks
# if the data are pandas series/dataframe, you need to first convert them into numpy arrays.

# It would be problematic to feed into a neural network values that all take wildly different ranges.
# This would make learning difficult.
# A commonly used practice is to do feature-wise normalization.

mean = train_data.mean(axis=0)
train_data -= mean
std = train_data.std(axis=0)
train_data /= std

test_data -= mean
test_data /= std

# Note that the quantities used for normalizing the test data are computed using the training data.
# You should never use in your workflow any quantity computed on the test data
```

```python
# we can now build the network.

model = tf.keras.models.Sequential([tf.keras.layers.Dense(64, activation=tf.nn.relu, input_shape=(train_data.shape[1],)), # first hidden layer
                                    tf.keras.layers.Dense(64, activation=tf.nn.relu),  # second hidden layer
                                    tf.keras.layers.Dense(1)]) # output layer


model.compile(optimizer='rmsprop', loss='mse', metrics=['mae'])
```

### Explanation of the neural network
- number of features in the input data = train_data.shape[1]
- for the output layer, we do not use any activation function. This is common for the regrssion task.
- the loss function is mse: mean squared error.
- the metrics is mae: mean absolute error, the mean absolute difference between the predicted housing price and actual the housing price.

```python
# we can fit the model
model.fit(train_data, train_targets,                                      
          epochs=80, batch_size=16, verbose=0)
# verbose = 0, it will not disply the loss and accuray values during training.
```

```python
model.evaluate(test_data, test_targets)
# we look at the second value, the mean absolute error in thousands.
```
